```python
import requests

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=INSIDER_TRANSACTIONS&symbol=IBM&apikey=demo'
r = requests.get(url)
data = r.json()

print(data)
```

```python
# Convert JSON data to a pandas DataFrame
if "data" in data:  # Check if the "data" key exists in the JSON response
    df = pd.DataFrame(data["data"])  # Extract the "data" key and convert it to a DataFrame

    # Save the DataFrame to a CSV file
    df.to_csv("IBM_insider_transactions.csv", index=False)
    print("Data saved to IBM_insider_transactions.csv")
else:
    print("No 'data' key found in the JSON response:", data)
```
